```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
danger_drop_speeds = [15, 13, 145, 137, 129, 113, 113, 105, 89, 97, 113, 105, 81]
railgun_speeds = [121, 113, 97, 105, 97, 73, 89, 81, 64, 64, 56, 56, 48]
combined_list = []
combined_list.append(railgun_speeds)
combined_list.append(danger_drop_speeds)
combined_list
```

```python
# Plot the charts and apply some styling
plt.plot(time, danger_drop_speeds, label = 'Danger Drop', color = 'red', marker = 'x')
plt.plot(time, railgun_speeds, label = 'RailGun', color = 'blue', marker = '^')
plt.grid()
plt.legend(['Danger Drop', 'RailGun'] , loc = 4)
plt.title('Coaster Speed Over Time')
plt.ylabel('Speed (km/hr)')
plt.xlabel('time')
plt.xlim = time.max()
plt.ylim = max(combined_list)
plt.savefig('Coaster Speed Over Time')
```
