```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
```

```python
cities_loc = [value for value in x_axis]
cities_loc
```

```python
# Create a bar chart based upon the above data
plt.xticks(cities_loc, cities)
plt.bar(x_axis, cars_in_cities, color = 'blue', alpha = 0.25)
plt.ylim(0, 600)
plt.ylabel('Commuting Cars Per 1, 000 Population Age 16+')
plt.xlabel('Cities')
```
