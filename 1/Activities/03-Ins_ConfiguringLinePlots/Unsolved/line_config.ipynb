```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis and variables
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
# Draw a horizontal line with 0.25 transparency
plt.plot(x_axis, sin)
plt.plot(x_axis, cos)
plt.hlines(0, x_axis.min(), x_axis.max(), alpha = 0.25)
```

```python
# Assign plots to tuples that stores result of plot

# Each point on the sine chart is marked by a blue circle
plt.plot(x_axis, sin, color = 'blue', label = 'sine', marker = 'o')

# Each point on the cosine chart is marked by a red triangle
plt.plot(x_axis, cos, color = 'red', label = 'cosine', marker = '^')
```

```python
# Adds a legend and sets its location to the lower right
plt.plot(x_axis, sin, color = 'blue', label = 'sine', marker = 'o')
plt.plot(x_axis, cos, color = 'red', label = 'cosine', marker = '^')
plt.legend(['Sine', 'Cosine'], loc = 4)
plt.savefig('sine_cosine_legend')
```
