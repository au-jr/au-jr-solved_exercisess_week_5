```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate the x values from 0 to 10 using a step of 0.1
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
# Add a semi-transparent horizontal line at y = 0
sin_plot = plt.plot(x_axis, sin, label = 'Sine', color = 'blue', marker = 's')
cos_plot = plt.plot(x_axis, cos, label = 'Cosine', color = 'red', marker = 'x')
plt.xlim(x_axis.min(), x_axis.max())
plt.ylim(-1, 1)
plt.hlines(0, x_axis.min(), x_axis.max(), alpha = 0.25)
plt.grid()
plt.savefig('aesthetic ;)')
```
