```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1, 13, 1)
x_axis
```

```python
# Avearge weather temp
points_C = [26.1, 25.3, 22.5, 20.6, 17.3, 14.5, 13.5, 15.8, 17.9, 19.2, 20.3, 24.7]
```

```python
# Convert to Fahrenheit F = C * 9/5 + 32
points_F = []
for c in points_C:
    F = c * 9/5 + 32
    F = round(F, 2)
    points_F.append(F)
points_F
```

```python
# Create a handle for each plot
celcius = plt.plot(x_axis, points_C, color = 'blue', marker = '+', label = 'Celcius')
farenheit = plt.plot(x_axis, points_F, color = 'red', marker = 's', label = 'Fahrenheit')
```

```python
# Set our legend to where the chart thinks is best
celcius = plt.plot(x_axis, points_C, color = 'blue', marker = '+', label = 'Celcius')
farenheit = plt.plot(x_axis, points_F, color = 'red', marker = 's', label = 'Fahrenheit')
plt.legend(['celcius', 'farenheit'], loc = 'best')

plt.xlabel('Month')
plt.ylabel('Temperature')
plt.savefig('temp_with_legend')
```
